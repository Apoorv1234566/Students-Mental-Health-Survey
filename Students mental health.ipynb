```python
df.columns = ['Timestamp', 'Gender', 'Age', 'Course', 'Year', 'CGPA', 'Marital Status', 'Depression', 'Anxiety', 'Panic Attack', 'Treatment']
```

```python
df.head()
```

```python
df['Year'].unique()
```

```python
df["Year"]=df["Year"].map({'Year 1':1,'year 1':1,'Year 2':2,'year 2':2,'Year 3':3,'year 3':3,'year 4':4})
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df['Timestamp'],errors='coerce')
df['day']=df['date'].dt.day
df['month']=df['date'].dt.month
df['year']=df['date'].dt.year
df['hour']=df['date'].dt.hour
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.year=df.year.astype(int)
df.day=df.day.astype(int)
df.month=df.month.astype(int)
df.hour=df.hour.astype(int)
df.Age=df.Age.astype(int)
```

```python
df["CGPA"].unique()
```

```python
df["CGPA"]=df["CGPA"].str.split("-").str[0]
```
